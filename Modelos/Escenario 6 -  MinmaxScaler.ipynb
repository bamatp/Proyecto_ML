# **ESCENARIO 6 - MIN MAX SCALER**

- Importamos el dataset del escenario 6

```python
import pandas as pd
Data_Science_Salaries = pd.read_csv("DS - Escenario 1.6.csv", sep = ",")
```

- Importamos las bibliotecas necesarias

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from lightgbm import LGBMRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.neighbors import KNeighborsRegressor
from catboost import CatBoostRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn import metrics
```

- Seleccionamos el target y las features del dataset

```python
X = Data_Science_Salaries.drop('salary_in_usd', axis=1)
y = Data_Science_Salaries['salary_in_usd']
```

```python
X.info()
```

```python
X.columns = X.columns.astype(str)
```

- Hacemos el x e y split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print("Total features shape:", X.shape)

print("Train features shape:", X_train.shape)
print("Train target shape:", y_train.shape)

print("Test features shape:", X_test.shape)
print("Test target shape:", y_test.shape)
```

- Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
predictions_lr = lr.predict(X_test)
```

```python
import numpy as np
mae_lr = metrics.mean_absolute_error(y_test,predictions_lr)
mse_lr = metrics.mean_squared_error(y_test,predictions_lr)
rmse_lr = np.sqrt(metrics.mean_squared_error(y_test,predictions_lr))
r2_lr = metrics.r2_score(y_test,predictions_lr)
mape_lr = metrics.mean_absolute_percentage_error(y_test,predictions_lr)
```

```python
print('MAE :'," ", mae_lr)
print('MSE :'," ", mse_lr)
print('RMSE :'," ", rmse_lr) 
print('R2:'," ", r2_lr) 
print('MAPE:'," ", mape_lr)
```

```python
linear_regression = {
    'Metric': ['MAE', 'MSE', 'RMSE', 'R2', 'MAPE'],
    'Value': [mae_lr, mse_lr, rmse_lr, r2_lr, mape_lr]
}
linear_regression_results = pd.DataFrame(linear_regression)
```

**GRIDSEARCH**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
from catboost import CatBoostRegressor
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

models = {

    'Catboost': (CatBoostRegressor(), {'learning_rate': [0.1, 0.01],'depth': [4, 6, 8],'l2_leaf_reg': [1, 3, 5],'iterations': [100, 200]})
}

for model_name, (model, param_grid) in models.items():
    print(f"Grid Search para el modelo {model_name}:")
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='r2')
    grid_search.fit(X_train, y_train)
    best_params = grid_search.best_params_
    best_score = grid_search.best_score_
    print(f"Mejores parámetros: {best_params}")
    print(f"Puntuación cruzada (cross-validation) del mejor modelo: {best_score}\n")
```

- Catboost

```python
cb2 = CatBoostRegressor(depth=6, iterations = 100, l2_leaf_reg= 5, learning_rate = 0.1)
cb2.fit(X_train,y_train)
predictions_cb2 = cb2.predict(X_test)

mae_cb2 = metrics.mean_absolute_error(y_test,predictions_cb2)
mse_cb2 = metrics.mean_squared_error(y_test,predictions_cb2)
rmse_cb2 = np.sqrt(metrics.mean_squared_error(y_test,predictions_cb2))
r2_cb2 = metrics.r2_score(y_test,predictions_cb2)
mape_cb2 = metrics.mean_absolute_percentage_error(y_test,predictions_cb2)
```

```python
print('MAE :'," ", mae_cb2)
print('MSE :'," ", mse_cb2)
print('RMSE :'," ", rmse_cb2) 
print('R2:'," ", r2_cb2) 
print('MAPE:'," ", mape_cb2)
```
