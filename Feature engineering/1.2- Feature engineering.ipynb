### Feature Engineering

```python
import pandas as pd
Data_Science_Salaries = pd.read_csv("ds_salaries.csv", encoding='latin-1')
```

```python
Data_Science_Salaries.head(5)
```

### Filtros

```python
Data_Science_Salaries = Data_Science_Salaries[(
    Data_Science_Salaries['employment_type'] == "FT")]
```

**1) Columna work_year**

```python
year_dummies = pd.get_dummies(Data_Science_Salaries['work_year'])
year_dummies
```

```python
Data_Science_Salaries=pd.concat([Data_Science_Salaries,year_dummies],axis=1)
```

```python
Data_Science_Salaries.drop(['work_year'],axis=1,inplace=True)
```

```python
Data_Science_Salaries.head()
```
